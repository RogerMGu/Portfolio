## Population by Continent
I'm going to group by continent and getting relevant and aggegated information

```python
continent_df = df[["2022 Population", "Area (km²)", "Growth Rate", "World Population Percentage"]].groupby(df["Continent"]).agg({ # or simply .sum()
    "2022 Population": 'sum',
    "Area (km²)": 'sum',
    "Growth Rate": 'mean',
    "World Population Percentage": 'sum'
})

continent_df["World Population Percentage 2"] = round(continent_df["2022 Population"] / continent_df["2022 Population"].sum() * 100, 2)
continent_df["Average Density (per km²)"] = round(continent_df["2022 Population"] / continent_df["Area (km²)"], 2)
continent_df = continent_df.sort_values("2022 Population", ascending=False)
continent_df
```

```python
continent_df = continent_df.drop("World Population Percentage", axis="columns")
continent_df = continent_df.rename(columns = {"World Population Percentage 2": 'World Population Percentage'})
continent_df
```

```python
continent_df.to_csv("continent_df.csv") # Saving the dataframe into a csv file.
```

Creating new columns with functions

```python
def totalGrowth(row):
    result = row["2022 Population"]/row["1970 Population"]
    return result
df["Total Growth since 1970"] = df.apply(totalGrowth, axis=1).round(4)

df.sort_values("Total Growth since 1970", ascending = False).head()
```
